```python
import pandas as pd

df1 = pd.read_excel("../data/bitcoin_merged.xlsx")
df2 = pd.read_csv('../../reddit/data/reddit_daily_normalized.csv')
df3 = pd.read_csv('../../reddit/data/news_daily_normalized.csv')
```

```python
df1 = df1[df1['Inflation Rate'].notna()]
```
